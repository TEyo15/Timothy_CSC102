#### PROJECT 1
#### PSEUDOCODE 1

<br> START
<br> INPUT list
<br> CONFIGURE bad word counter
<br> INPUT forbidden letters
<br> num = 0
<br> FOR word in list
   * FOR letter in word
       * IF letter in forbidden letters
           * DISPLAY "There are bad letters"
           * num += 1
       * ENDIF
       * DISPLAY word, "number of bad letters = ", num
       * APPEND num to bad word counter
   * ENDFOR
   
<br> DISPALY MIN (bad word counter)
<br> END

```python
#### CODE

list1 =["baby", "bag", "virus", "sport"]

bad_counter = []

print("The words are", end = "")
print(list1)
bad_letters = list(str(input("what are your forbidden letters (e.g \'e\')")))

def avoids(a):
    num = 0
    correct = True
    word = a
    for i in list(word):
        if i in bad_letters:
            # print("no")
            correct = False
            num += 1
        else:
            pass
            #print("yes")
    print(a, correct, "number of bad letters",num, end = "\n\n")
    bad_counter.append(num)
    
for a in list1:
    avoids(a)
    
hard = (min(bad_counter))

least = (bad_counter.index(hard))
print(list1[least], "has the least of bad lettes")
```
